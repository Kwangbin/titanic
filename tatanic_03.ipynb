##결측치
결측치는 비어있는 데이터를 의미합니다.

결측치에 대한 처리는 매우 중요합니다.

결측치에 대한 처리를 해주려면 다음의 내용을 반드시 알아야 합니다.

-결측 데이터 확인

-결측치가 아닌 데이터 확인

-결측 데이터 채우기

-결측 데이터 제거하

##결측치 확인 - isnull(), isnan()
컬럼(column)별 결측치의 갯수를 확인하기 위해서는 sum() 함수를 붙혀주면 됩니다.

sum()은 Pandas의 통계 관련 함수이며, 통계 관련 함수는 추후에 더 자세히 알아볼 예정입니다.

```python
df.isnull()
```

```python
df.isnull().sum()
```

##isna()

isnull() 과 동작이 완전 같습니다. 편한 것으로 써주세요. (심지어 도큐먼트도 같습니다)

```python
df.isna().sum()
```

#DataFrame 전체 결측 데이터의 갯수를 합산하기 위해서는 sum()을 두 번 사용하면 됩니다.

```python
df.isnull().sum().sum()
```

##결측치가 아닌 데이터 확인 - notnull()
notnull()은 isnull()과 정확히 반대 개념입니다.

```python
df.notnull()
```

```python
df.notnull().sum()
```

##결측 데이터 필터링
isnull() 함수가 결측 데이터를 찾는 boolean index 입니다.

즉, loc에 적용하여 조건 필터링을 걸 수 있습니다.

```python
df.loc[df['age'].isnull()]
```

##결측치 채우기 - fillna()
fillna()를 활용하면 결측치에 대하여 일괄적으로 값을 채울 수 있습니다.

```python
df1=df.copy()
```

```python
df1.tail()
```

```python
#888번 index의 결측치가 700으로 채움

df['age'].fillna(700).tail()
```

```python
df1.tail()
```

```python
df1['age']=df['age'].fillna(700)
```

```python
df1.tail()
```

카테고리 형 데이터을 채워주기 위해서는 다음과 같은 과정을 거쳐야 합니다.

이미 카테고리가 추가된 'A'나 'B'는 바로 fillna() 할 수 있습니다.

```python
df1['deck']
```

```python
df1['deck'].fillna('A')
```

```python
# 카테고리 목록에 없으면 않됨
df1['deck'].fillna('No Data')
```

```python
#카테고리로 채워주고자 할 때는 먼저 add_categories로 카테고리를 추가한 후 채워야 합니다.
df1['deck'].cat.add_categories('No Data').fillna('No Data')
```

##통계값으로 채우기

```python
df2=df.copy()
```

```python
df2.tail()
```

```python
# 평균으로 채우기

df2['age'].fillna(df2['age'].mean()).tail()
```

```python
# 중앙값ㅇ 로 채우기

df2['age'].fillna(df2['age'].median()).tail()
```

```python
#최빈값으로 채우기

df2['deck'].mode()
```

```python
df2['deck'].value_counts()
```

```python
# 최빈값(mode)으로 채울 때에는 반드시 0번째 index 지정하여 값을 추출한 후 채워야 합니다.
df2['deck'].mode()
```

```python
# 인덱스 지정이 없으면 채워 지지가 않는다,,,
df2['deck'].fillna(df2['deck'].mode()).tail()
```

```python
# index 지정
df2['deck'].mode()[0]

df2['deck'].fillna(df2['deck'].mode()).tail()
```

```python
# index [0]   까지 지정 해야 함

df2['deck'].fillna(df2['deck'].mode()[0]).tail()
```

##NaN 값이 있는 데이터 제거하기 (dropna)

```python
df3 = df.copy()
```

```python
df3.tail()
```

```python
#dropna()로 1개 라도 NaN 값이 있는 행은 제거할 수 있다. (how='any')

df3.dropna()
```

###기본 옵션 값은 how=any로 설정되어 있으며, 다음과 같이 변경할 수 있습니다.

any: 1개 라도 NaN값이 존재시 drop

all: 모두 NaN값이 존재시 drop

```python
df3.dropna(how='all')
```
